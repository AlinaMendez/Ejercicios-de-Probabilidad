### Exercise 1

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import random

#Funtion to roll the dice.
dado_1 = random.choices([1,2,3,4,5,6],k=1000)
dado_2 = random.choices([1,2,3,4,5,6],k=1000)

dado = dado_1 + dado_2
num = 0
def suma(x):
    num = 0
    for i in range(1000):
        if x[i]%2 == 0 or x[i] >= 7:
            num += 1
    return(num)

probabilidad = suma(dado)/1000
print(probabilidad)
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import numpy as np

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  
  
#Initialize important variables

cantidad_red = 0
cantidad_white = 0
cantidad_green = 0
opcion_1 = 0
opcion_2 = 0
n_simulations = 1000000

for i in range(n_simulations):
    
    #make a list of the colors that we choose
   
    list = random.choices(d,k = 5)

    #Convert it to a numpy array
    eleccion = np.array(list)

    #find the number of each taht we picked
    
    for i in range(len(eleccion)):
        if eleccion[i] == 'White':
            cantidad_white += 1
        elif eleccion[i] == 'Red':
            cantidad_red += 1
        else: 
            cantidad_green += 1
    
    #Keep track if the combination met the criteria

    if cantidad_white == 3 or cantidad_red == 2:
        opcion_1 += 1
    elif cantidad_white == 5 or cantidad_red == 5 or cantidad_green ==5:
        opcion_2 += 1
        
    
#print 'the probability of 3 white and 2 red is' in percentage
print("The probability of 3 white and 2 red is "+str(opcion_1*100/n_simulations))
print("Cantidad de veces que salen 3 white and 2 red "+ str(opcion_1)+ " en "+ str(n_simulations)+ " simulaciones")

#print 'the probability of all the same color is' in percentage   
print("the probability of all the same color is "+str(opcion_1*100/n_simulations))
print("Cantidad de veces que salen todas del mismo color "+str(opcion_2)+ " en "+ str(n_simulations)+ " simulaciones")
```
